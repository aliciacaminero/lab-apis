```python
"""
1. Lista de tus artistas favoritos**:
    - Haz una lista de tus tres artistas favoritos y guárdala en una variable llamada `artists`.
    - Ejemplo: `artistas = [«Los Fabulosos Cadillacs», «Manu Chao», «Muchachito Bombo Infierno»]`.
"""
```

```python
artists = ["Coldplay", "One Republic", "Imagine Dragons", "Recycled J", "Natos y Waor"]
```

```python
""" 
2. Obtener las mejores canciones**:
    - Escribe una función llamada `get_top_tracks`.
    - Esta función debe aceptar el nombre de un artista y devolver el nombre de las 5 primeras canciones de ese artista.
    - Utiliza la función `get_top_tracks` para obtener las 5 primeras canciones de cada artista de tu lista `artists` y guarda los resultados en una nueva lista llamada `top_tracks_list`.
"""
```

```python
def get_top_tracks(artist_name):
    results = sp.search(q='artist:' + artist_name, type='artist', limit=1)
    artist = results['artists']['items']

    if artist:
        artist_id = artist[0]['id']
        # Obtener las canciones más populares del artista
        top_tracks = sp.artist_top_tracks(artist_id)
        # Devolver solo los nombres de las 5 primeras canciones
        return [track['name'] for track in top_tracks['tracks'][:5]]
    return []
```

```python
top_tracks_list = {artist: get_top_tracks(artist) for artist in artists}

for artist, tracks in top_tracks_list.items():
    print(f"{artist}: {tracks}")
```

```python
""" 
3. Descubrir artistas relacionados**:
    - Escribe una función llamada `find_related_artists`.
    - Esta función debe aceptar el nombre de un artista y devolver los nombres de los 5 primeros artistas relacionados con el artista proporcionado.
    - Almacena los resultados en una lista llamada `related_artists_list`.
"""
```

```python
# Función que encuentra los artistas relacionados
def find_related_artists(artist_name):
    try:
        # Buscar el artista
        results = sp.search(q='artist:' + artist_name, type='artist', limit=1)
        artist = results['artists']['items']

        if artist:
            artist_id = artist[0]['id']
            # Obtener los artistas relacionados
            related_artists = sp.artist_related_artists(artist_id)
            # Devolver solo los nombres de los 5 primeros artistas relacionados
            return [related['name'] for related in related_artists['artists'][:5]]
        else:
            print(f"No se encontró el artista: {artist_name}")
            return []  # Si no hay resultados, retornamos una lista vacía.
        
    except Exception as e:
        print(f"Error al buscar artistas relacionados para {artist_name}: {e}")
        return []  # En caso de error, retornamos una lista vacía.   

# Lista de artistas para los que queremos encontrar artistas relacionados -> artist

# Almacenar resultados en related_artists_list
related_artists_list = []

for artist in artists:
    related_artists = find_related_artists(artist)
    related_artists_list.append({artist: related_artists})

# Imprimir los resultados
for item in related_artists_list:
    for artist, related in item.items():
        if related:  # Solo imprimir si hay artistas relacionados
            print(f"{artist}: {', '.join(related)}")
        else:
            print(f"{artist}: No se encontraron artistas relacionados.")
        #print(f"{artist}: {related}")
```

### Getting a Playlist's Details
To fetch details about a specific playlist, you can use the playlist method. You'll need the playlist's Spotify ID.

In this example, we will use the following playlist id: *37i9dQZF1DXd9zR7tdziuQ*

```python
playlist_id = "37i9dQZF1DXd9zR7tdziuQ"
playlist = sp.playlist(playlist_id)
```

```python
print(playlist['name'])  # Print the playlist's name
print(playlist['description'])  # Print the playlist's description
```

### Getting Tracks from a Playlist
If you want to get the tracks from a specific playlist, you can use the playlist_tracks method.

```python
tracks = sp.playlist_tracks(playlist_id)
for track in tracks['items']:
    print(track['track']['name'])  # Print each track's name
```

### Getting Artists from a Playlist

To extract all the artists from the tracks in a playlist, you'd typically follow these steps:

1. Fetch the playlist's tracks.
2. Iterate through each track.
3. For each track, extract the associated artists.

```python
# List to store unique artist names
artists_list = []

for track_item in tracks['items']:
    track = track_item['track']
    for artist in track['artists']:
        artist_name = artist['name']
        if artist_name not in artists_list:  # This ensures each artist is added only once
            artists_list.append(artist_name)

print(artists_list)
```

```python

# Función para obtener y mostrar las listas de reproducción destacadas
def get_featured_playlists():
    try:
        # Obtener listas de reproducción destacadas
        featured_playlists = sp.featured_playlists(country='US', limit=5)  # Aquí 'country' es opcional

        # Mostrar los 5 primeros nombres e IDs de las listas de reproducción destacadas
        print("Las 5 listas de reproducción más destacadas son:")
        for idx, playlist in enumerate(featured_playlists['playlists']['items'], start=1):
            name = playlist['name']
            playlist_id = playlist['id']
            print(f"{idx}. Nombre: {name} | ID: {playlist_id}")
    
    except Exception as e:
        print(f"Error al obtener las listas de reproducción destacadas: {e}")

# Llamar a la función
get_featured_playlists()
```

```python
""" 
2. Profundizar:
   - Elige una de las 5 listas de reproducción destacadas (puedes elegir la que te parezca más interesante o simplemente elegir una al azar).
   - Obtén y muestra su nombre, descripción y número total de pistas.
"""
```

```python
# Función para obtener y mostrar detalles de una lista de reproducción específica
def get_playlist_details(playlist_id):
    try:
        # Obtener detalles de la lista de reproducción
        playlist = sp.playlist(playlist_id)

        # Extraer nombre, descripción y número total de pistas
        name = playlist['name']
        description = playlist['description']
        total_tracks = playlist['tracks']['total']

        # Mostrar los detalles
        print(f"Nombre: {name}")
        print(f"Descripción: {description}")
        print(f"Número total de pistas: {total_tracks}")

    except Exception as e:
        print(f"Error al obtener los detalles de la lista de reproducción: {e}")

# Elegir una lista de reproducción (por ejemplo, 'Top 50' con ID: 37i9dQZEVXbMDoHDwVN2tF)
# Puedes elegir otra ID de las 5 listas que obtuviste previamente
playlist_id = "37i9dQZEVXbMDoHDwVN2tF"  # Ejemplo de ID para "Top 50"

# Llamar a la función para obtener los detalles de la lista de reproducción seleccionada
get_playlist_details(playlist_id)
```

```python
""" 
3. Pista-tástica:
   - Extrae y muestra los nombres de las 10 primeras pistas de la lista de reproducción elegida.
"""
```

```python
def get_first_10_tracks(playlist_id):
    try:
        # Obtener las primeras 10 pistas de la lista de reproducción
        tracks = sp.playlist_tracks(playlist_id, limit=10)

        # Mostrar los nombres de las primeras 10 pistas
        print("Las primeras 10 pistas de la lista de reproducción son:")
        for idx, item in enumerate(tracks['items'], start=1):
            track_name = item['track']['name']
            artist_name = item['track']['artists'][0]['name']  # Obtener el primer artista
            print(f"{idx}. {track_name} - {artist_name}")

    except Exception as e:
        print(f"Error al obtener las pistas: {e}")

# Elegir una lista de reproducción (por ejemplo, 'Top 50' con ID: 37i9dQZEVXbMDoHDwVN2tF)
playlist_id = "37i9dQZEVXbMDoHDwVN2tF"  # Ejemplo de ID para "Top 50"

# Llamar a la función para obtener las primeras 10 pistas
get_first_10_tracks(playlist_id)
```

```python
""" 
4. Estilo artístico:
   - Crear un diccionario donde las claves sean los nombres de las 10 primeras pistas, y los valores sean listas que contengan los nombres de los artistas asociados a cada pista.
   - Por ejemplo: `{«TrackName1»: [«Artista1», «Artista2»], «TrackName2»: [«Artista3»]}`
"""
```

```python
# Función para obtener las primeras 10 pistas y crear un diccionario con los artistas
def get_tracks_artists_dict(playlist_id):
    try:
        # Obtener las primeras 10 pistas de la lista de reproducción
        tracks = sp.playlist_tracks(playlist_id, limit=10)

        # Crear un diccionario para almacenar el nombre de la pista y los artistas
        tracks_artists_dict = {}

        # Iterar sobre las pistas y agregar al diccionario
        for item in tracks['items']:
            track_name = item['track']['name']
            # Obtener los nombres de los artistas para cada pista
            artists = [artist['name'] for artist in item['track']['artists']]
            # Agregar al diccionario
            tracks_artists_dict[track_name] = artists

        # Mostrar el diccionario
        print("Diccionario de pistas y artistas:")
        for track, artists in tracks_artists_dict.items():
            print(f"«{track}»: {artists}")
    
    except Exception as e:
        print(f"Error al obtener las pistas y los artistas: {e}")

# Elegir una lista de reproducción (por ejemplo, 'Top 50' con ID: 37i9dQZEVXbMDoHDwVN2tF)
playlist_id = "37i9dQZEVXbMDoHDwVN2tF"  # Ejemplo de ID para "Top 50"

# Llamar a la función para obtener el diccionario de pistas y artistas
get_tracks_artists_dict(playlist_id)
```
